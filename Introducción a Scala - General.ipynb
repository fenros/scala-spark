### Conceptos básicos:

* El nombre _Scala_ proviene de "Scalable Language" (se adapta tanto a pequeños scripts como a grandes sistemas)
* Se ejecuta en la JVM y es compatible con las librerías Java (e incluso puede convivir con código Java)
* Es un lenguaje estáticamente tipado (aunque no siempre es necesario declarar el tipo)
* Mezcla la orientación a objetos con la programación funcional (aunque no obliga a su aplicación "pura")
* Elimina código innecesario generando programas más cortos (principalmente respecto a Java)

### Objetivos de este notebook:

* Repasar conceptos básicos sobre programación funcional y orientación a objetos
* Practicar la sintaxis del lenguaje
* Reforzar las habilidades para programar de manera funcional frente a la imperativa
* Conocer algunas características 'avanzadas' de Scala

Resumiendo, aprenderemos a usar Scala de manera gradual y progresiva, evolucionando desde los conceptos más básicos hasta algunas de las funcionalidades más avanzadas del lenguaje, las cuales nos permitirán desarrollar aplicaciones definiendo elementos pequeños para ir construyendo con ellos otros más grandes.
